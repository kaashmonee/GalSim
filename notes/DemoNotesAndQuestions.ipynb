# demo3.py

- Supposed to encompass main features of
  an image from a ground based telescope.
- Galaxy : sum of a bulge and a disk
    * (Q) What are the technical definitions for a
          bulge and a disk?
    * each component represented by a sheared Sersic profile.
    * (Q) What is a "sersic profile?"
    * (A) A "sersic" profile (https://en.wikipedia.org/wiki/Sersic_profile): 
        - Mathematical function that describes how intensity $I$ of a galaxy varies with distance $R$ from center.
        - Most galaxies are fit by Sersic profiles with indices that range from $1/2 < n < 10$.
        - $n=4$ good approximation for elliptical galaxies
        - $n=1$ good approximation for spiral disks
        - There's a correlation with sersic indices and masses of supermassive black holes at the center of a galaxy...
- PSF: contains atmospheric and optical components
- Atomspheric component is a "Kolmogorov turbulent spectrum"
    * (Q) What is a "Kolmogorov turbulent spectrum?" Why is that used here?
- Optical component includes "defocus, coma, and astigmatism" 
    * (Q) What are defocus, coma, and astigmatism?
    * (A) 
        - Defocus: a type of aberration where the image is out of focus.
        - Coma: A type of aberration that is present in telescopes with parabolic mirrors where if viewed from an angle that is not the focal point of the parabolic mirror, the image appears sheared/distorted (sometimes resulting in a coma-like tail) --- in other words, "comatic," which is it's called a "comatic" aberration or a coma.
        - Astigmatism (http://hyperphysics.phy-astr.gsu.edu/hbase/geoopt/coma.html): This happens when at different planes of the lens the image is not focused onto a specific point on the other side. From what I understand, this occurs when the curvature of the lens is different from the desired curvature at that point/plane.
  as well as "obscuration from a secondary mirror."
    * (Q) What is obscuration from a secondary mirror?
    * (A) Just what it sounds like. Potential sources of aberration from a 2nd mirror?
- Adds poisson and gaussian read noise
        * (Q) What is "read noise" and why add both kinds of noise?
        * (A) Poisson shot noise or read noise occurs from random electronic signals resulting in false positives on the CCD. For large $N$, or large signals, the noise becomes Gaussian in nature.
- Lets the pixels be "distorted in the sky"
    * (Q) How does it do this?
    * (A) The image is sheared.

```python
demo3img = fits.open(images_path+"demo3.fits")[0].data
demo3psf_img = fits.open(images_path+"demo3_epsf.fits")[0].data
demo3opticalpsf = fits.open(images_path+"demo3_opticalpsf.fits")[0].data

for img in [demo3img, demo3psf_img, demo3opticalpsf]:
    plt.figure()
    plt.imshow(img, cmap="gray")
```
